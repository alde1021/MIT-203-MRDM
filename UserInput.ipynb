```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/mlula/Downloads/MRDM/loan_dataset_50rows.csv')
df
```

```python
X = df[['Income','CreditScore','PreviousLoans']]
y = df['LoanApproved']
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
income = int(input("Enter Income: "))
credit = int(input("Enter Credit Score: "))
loans = int(input("Enter Previous Loans: "))

result = model.predict([[income, credit, loans]])

print("\nLoan Evaluation Result:")

if result[0] == 1:
    print("✅ Loan Approved")
else:
    print("❌ Loan Rejected")
```
